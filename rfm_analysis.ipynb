```python
import pandas as pd
```

```python
# Load Data
df = pd.read_csv("C:\data\single_customer_view.csv")
```

```python
df.head()
```

```python
# Date Conversion
df['LastPurchaseDate'] = pd.to_datetime(df['LastPurchaseDate'])
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# Recency Calculation
today = df['LastPurchaseDate'].max() + pd.Timedelta(days=1)

df['Recency'] = (today - df['LastPurchaseDate']).dt.days
```

```python
df[['Customer_name','Recency']].head()
```

```python
# Frequency and Monetary Calculation
df['Frequency'] = df['TotalOrders']
df['Monetary'] = df['TotalSales']
```

```python
df[['Customer_name','Frequency','Monetary']].head()
```

```python
df[['Customer_name','Recency','Frequency','Monetary']].head()
```

```python
def segment_customer(row):
    if row['Recency'] <=2 and row['Frequency'] >=25 and row['Monetary'] >=600000:
        return 'Champion'
    elif row['Frequency'] >=20 and row['Monetary'] >=300000:
        return 'Loyal'
    else:
        return 'Hibernating'

df['Segment'] = df.apply(segment_customer, axis=1)
```

```python
df[['Customer_name','Segment']].head()
```

```python
df['Segment'].value_counts()
```

```python
df.to_csv("rfm_output_consumer360.csv", index=False)
```
